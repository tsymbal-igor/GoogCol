```python
import numpy as np
import random as rnd
from collections import namedtuple, defaultdict
from IPython.display import clear_output
import time
```

```python
#Coord = namedtuple('Coord', ['x', 'y'])
                
WIDTH = 15
HIGHT = 5

WAIT_TIME = 0.3

maze = np.zeros([HIGHT,WIDTH])

route = []
x, y = 0, 0

class Cell():
    def __init__(self, y, x, parent=0, visited = False):
        self.x, self.y = x, y
        self.visited = visited
        self.parent = parent

def free_directions(y, x, maze):
    _dir = []
    
    if y > 0 and maze[y - 1, x]<1:
        _dir.append('u')
    
    if x < (maze.shape[1] - 1) and maze[y, x + 1]<1:
        _dir.append('r')
    
    if y < (maze.shape[0] - 1) and maze[y + 1, x]<1:
        _dir.append('d')
    
    if x > 0 and maze[y, x - 1]<1:
        _dir.append('l')
    
    return _dir

def next_direction(_dir):
    return rnd.choice(_dir)

route = [Cell(x, y)]

i = 0
child_index = 1
parent_index = 0

step = WIDTH*HIGHT

run_flag = True

while run_flag:
    print('\n===============================\n''шаг цикла', i,
          '\n индекс родителя', parent_index,
          '\n индекс потомка', child_index)
 
    y, x = route[parent_index].y, route[parent_index].x 
    
    maze[y, x] = len(route)
    
    if len(route)==step:
        run_flag=False
    
    d = free_directions(y,x,maze)
    print('возможный направления', d)
    if d!=[]:
        s = next_direction(d)

        print(' направление', s)
    
        x_next = 0
        y_next = 0
    
        if s == 'u':
            x_next = x + 0
            y_next = y + -1
        if s == 'r':
            x_next = x + 1
            y_next = y + 0
        if s == 'd':
            x_next = x + 0
            y_next = y + 1
        if s == 'l':
            x_next = x - 1
            y_next = y + 0    
        
        print("\nтекущее положение \n (y, x) = (%d, %d) \n колво пройденых ячеек %d \n \n следующее положение \n (y, x) = (%d, %d)"%(y, x, len(route), y_next, x_next))
        
        #maze[y, x] = len(route)
        
        route.append(Cell(y_next, x_next, parent_index, visited = True))
        parent_index=child_index
        child_index+=1
    else:
        parent_index = route[parent_index].parent 
        print('\n \n \n \n \n \n \n')
    #i+=1
    
    print('\n',maze)
    #input()
    time.sleep(WAIT_TIME)
    clear_output(wait=True)
print("The end, world!")    
```
